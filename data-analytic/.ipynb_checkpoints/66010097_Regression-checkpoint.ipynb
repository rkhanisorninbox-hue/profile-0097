```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("HR_Analytics.csv.csv")
print("ตัวอย่างข้อมูล:")
display(data.head())
```

```python
print("\nCorrelation Matrix:")
display(data.corr(numeric_only=True))
```

(เลือก feature + target)

```python
X = data[["Age", "DistanceFromHome", "JobLevel"]].values
y = data["MonthlyIncome"].values
X = np.c_[np.ones(X.shape[0]), X]
```

Gradient Descent

```python
def gradient_descent(X, y, alpha=0.001, epochs=1000):
    m, n = X.shape
    theta = np.zeros(n)
    cost_history = []

    for i in range(epochs):
        y_pred = X.dot(theta)
        error = y_pred - y
        cost = (1/(2*m)) * np.dot(error.T, error)
        cost_history.append(cost)
        gradient = (1/m) * X.T.dot(error)
        theta -= alpha * gradient
    
    return theta, cost_history

theta, cost_history = gradient_descent(X, y, alpha=0.001, epochs=2000)

print("\nParameter ที่ได้จาก Gradient Descent:", theta)
```

Cost Function

```python
plt.plot(range(len(cost_history)), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Cost Function")
plt.show()
```
