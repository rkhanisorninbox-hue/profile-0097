2.4 การเลือกตามเงื่อนไข (Conditional Selection)
	จากการทำการ indexing นั้น เราสามารถนำผลลัพธ์ที่ได้จากการเลือกแอททริบิวต์ที่สนใจ มาทำการตรวจสอบเงื่อนไขได้ ตัวอย่าง เช่น หากต้องการตรวจสอบรายละเอียดไวน์ที่ผลิตจาก Italy

```python
df.country == 'Italy'
```

การใช้ Operation แบบนี้แสดงออกมาเป็น Boolean (True/False) หมายถึง แต่ละแถวมี Italy หรือไม่ ? 
 Operation นี้ จะใช้ตรวจสอบว่าแต่ละ record หรือแต่ละแถว เป็น Italy หรือไม่ด้วย loc 

```python
df.loc[df.country == 'Italy']
```

เราสามารถใช้คำสั่ง and ด้วยเครื่องหมาย ampersand ( & ) หรือคำสั่ง or ด้วยเครื่องหมาย pipe ( | ) ได้
เช่น หากต้องการทราบไวน์อิตาลีและได้คะแนน (points) มากกว่า 90 จะใช้คำสั่ง

```python
df.loc[(df.country == 'Italy') & (df.points >=90)]
```

หากต้องการทราบไวน์อิตาลี หรือ ไวน์ที่ได้คะแนน (points) มากกว่าหรือเท่ากับ 90 จะใช้คำสั่ง

```python
df.loc[(df.country == 'Italy') | (df.points >=90)]
```

คำสั่ง Maps ใช้สำหรับในการ Transformation ข้อมูล เช่นต้องการทราบคะแนน (point) ของไวน์ที่ได้รับว่าต่างจาก mean แค่ไหน เช่น 

```python
df.taster_name.unique()
```

```python
review_points_mean = df.points.mean()
df.points.map(lambda p: p - review_points_mean)
```

คำสั่ง .apply() จะใช้ในการแปลงข้อมูลในแต่ละคอลัมน์ 

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

df.apply(remean_points, axis='columns')
```

4. Grouping and Sorting
4.1 Groupwise 
	เนื่องจาก Data Frame เป็นข้อมูลแบบสองมิติที่อาจมีมากกว่าหนึ่งคอลัมน์ บางครั้งอาจมีบางคอลัมน์ที่ข้อมูลซ้ำซ้อนกัน ซึ่งสามารถนำมาจัดกลุ่มได้ และอาจใช้มากกว่าหนึ่งคอลัมน์ โดยใช้ฟังก์ชัน groupby() ในการจัดกลุ่ม และยังสามารถตามด้วยฟังก์ชันทางสถิติต่าง ๆ (sum(), count(), mean(), min(), max() ) เพื่อให้ได้ผลลัพธ์ตามที่ต้องการ
ตัวอย่าง จากคำสั่ง groupby() และ value_counts
Groupby จะใช้สร้างกลุ่มของ review ที่มีคะแนนเหมือนกัน จากนั้นในแต่ละ group จะแสดงว่ามีจำนวนเท่าไหร่ที่มีคะแนนเหมือนกันในอีกคอลัมน์ หนึ่ง
เช่นเดียวกัน สามารถจับกลุ่ม เช่น หาราคาไวน์ที่ถูกที่สุดในแต่ละคะแนน (points)


```python
df.groupby('points').points.count()
```

```python
df.groupby('points').price.min()
```

```python
df.groupby('country').price.min()
```

```python
df.groupby(['country','points']).price.min()
```

```python
#คำสั่ง .agg() ทำให้เรียกใช้ฟังก์ชันต่าง ๆ เช่น จำนวน ค่าน้อยสุด มากสุด ได้ เช่น 
df.groupby(['country']).price.agg([len, min, max])
```

```python
countries_reviewed = df.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

4.2 การจัดเรียงข้อมูล (Sorting)
	ข้อมูลจากการทำ groupby อาจไม่ได้ทำการเรียงลำดับมาก่อน การใช้คำสั่ง sort_values() จะช่วยในการเรียงลำดับ ได้ เช่น 

```python
#SORTING
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

ปกติแล้ว default จะเป็นแบบ ascending (น้อยไปมาก) เราสามารถเปลี่ยนเป็น descending (มากไปน้อย) ได้โดยชึ้คำสั่ง ascending = False ตัวอย่าง เช่น

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
#การ Sort ตาม index สามารถทำได้ ด้วยคำสั่ง sort_index() 
countries_reviewed.sort_index()
```

```python
#และยังสามารถ sort ได้มากกว่าหนึ่งคอลัมน์ เช่น ต้องการเรียงลำดับตามประเทศ (country) และ len
countries_reviewed.sort_values(by=['country', 'len'])
```
