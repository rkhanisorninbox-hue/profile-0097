# Broadcasting
เมื่อ มิติ (dimension) ของ array ที่จะเอามาทำ operation กันไม่เท่ากัน 
โดยจะทำ broadcasting ได้ array ทั้งสองตัวจะต้องมีคุณสมบัติข้อใดข้องหนึ่งดังนี้
1.ขนาดและรูปร่างของ dimension เท่ากันทั้งสอง array
2.ขนาดของ dimension ของ array ตัวใดตัวหนึ่งมีค่าเป็น 1
3.ถ้ามิติไม่เท่ากัน ขนาดของ dimension ของ array มีค่าเท่ากันหรือมีค่าเป็น 1

เช็ค compatibility ของ array dimensions ดูจาก dimension ขวาสุด แล้วทำจากขวาไปซ้าย

A =  3   x   5
B =  3   x   8
A x B = error

A =  	2   x   2  x  3
B =                 2  x  3
A x B = 2 x 2 x 3

# Dyanamic chart
Python ทั่วไปกับ Jupyter ไม่เหมือนกัน
ทอยลูกเต๋า 1 ลูก ถ้าทำแบบ unbias แล้วยิ่งทอยมากครั้ง หน้าที่ออกมันจะวิ่งเข้าหาจุดที่มีจำนวนความถี่เท่าๆกัน
ทอยลูกเต๋า 2 ลูก ถ้าทำแบบ unbias หน้าที่ออกจะมีความถี่ที่เลข 7 มากที่สุด
สีของ chart โดย default จะเปลี่ยนไปเรื่อยๆ 
Histogram เหมาะกันข้อมูล continuous แต่ข้อมูลหน้าลูกเต๋าเป็นข้อมูลไม่ต่อเนื่องแบบ discrete เพราะมีแค่ 1 2 3 4 5 6  จะเห็นว่างช่วงมันมีความห่างในแกน x ไม่เท่ากัน 

```python
def roll_dice(num_dices, size):
    rng = np.random.default_rng()
# roll the dice
    faces = rng.integers(low=1, high=6, endpoint=True, size=(size,num_dices))
# sum face values
    return np.sum(faces, axis=1)
```

```python
# Roll 1 dice
from IPython import display
# Create empty zero-sized numpy array
face_counts = np.array([], dtype=int)
for i in range(0,100):
    # RoLL 1 dice
    dice_faces = roll_dice(1,100)
    # Concatenate the new result to the previous results
    face_counts = np.concatenate((face_counts, dice_faces), axis=0)
    # Plot histogram of all results
    hist = plt.hist(face_counts)
    # CLear Jupyter display output
    display.clear_output(wait=True)
    # Display the active figure
    display.display(plt.gcf())
    
display.clear_output(wait=True)
```

```python
#Roll 2 dices

from IPython import display
# Create empty zero-sized numpy array
face_counts = np.array([], dtype=int)
for i in range(0,100):
    # Roll 2 dice
    dice_faces = roll_dice(2,100)
    # Concatenate the new result to the previous results
    face_counts = np.concatenate((face_counts, dice_faces), axis=0)
    # Plot histogram of all results
    hist = plt.hist(face_counts, bins=np.arange(1,14)-0.5, color = 'blue', rwidth= 0.8)
    # CLear Jupyter display output
    display.clear_output(wait=True)
    # Display the active figure
    display.display(plt.gcf())
    
display.clear_output(wait=True)
```

# Dynamic chart
ใช้กับการเรียกข้อมูลจากฐานข้อมูลเช่น firebase มาแสดงผล เช่น เก็บข้อมูล อุณหภูมิกับเวลาไว้ อยากเรียกมาแสดงผลค่า Temp กับ เวลาในช่วง 10 นาที โดยให้แสดงผลค่าเฉลี่ย Temp ภายใน 10 นาทีนี้ด้วย

```python
สร้างตัวแปร array => x, y, accum (สะสมค่า temp เพื่อเอาไว้หาค่า mean)
สร้างตัวแปร count = 1 และ max_count = 10 เอาไว้เพื่อต้องการแสดงแค่ 10 ค่า ปัจจุบัน
สร้าง figure ขนาด 6 * 4 
ใส่ค่าแกน ylim ที่ 20 - 50, label แกน x เป็นเวลา แกน y เป็น temp
```

```python
import numpy as np
from numpy.random import default_rng
import matplotlib.pyplot as plt
from IPython import display
import time

# Create random number generator
rng = default_rng()
x = np.array([], dtype=int)
y = np.array([], dtype=int)
accum_data = np.array([], dtype=int)
#นับว่าเกิน 10 ค่าข้อมูลหรือยัง
count = 1
max_count = 10
# สร้าง fig
fig = plt.figure(figsize=(6,4))
plt.ylim(20,50)
plt.xlabel('time(min)')
plt.ylabel('Temp Value')
```

```python
วนลูป for Gen temp data จำนวน 50 ค่า 
random ค่า temp ระหว่างช่วง 30 -40 องศา
ค่า x เก็บเวลา ค่า y เก็บ temp โดยทำ concatenate ค่าที่ random ได้ไปเรื่อยๆ
เก็บค่า y ใน accum เพื่อเอาไปหาค่าเฉลี่ย
เช็คว่าค่า > 10 รึยัง ถ้ามากกว่า 10 ต้อง slide เอาค่าภายใน 10 มาแสดงผล
เพิ่ม counter
หาค่า mean
แสดงค่า mean ที่ตำแหน่ง title
plot chart
รอ 1 วินาทีก่อนวนลูป gen temp data ครั้งต่อไป
```

```python
for i in range(0,50):
    #gen temp 30 - 40 degree
    data = rng.choice(np.arange(30,40))
    # Add a new data to x and y array
    x = np.concatenate((x, np.array([i])))
    y = np.concatenate((y,np.array([data])))
    
    #accum เพื่อหาค่า mean
    accum_data =np.concatenate((accum_data, np.array([data])))
    
    # Check if we need to shift x-axis series
    if count > max_count:
        x = x[1:]
        y = y[1:]
        plt.xlim(np.min(x), np.max(x))
        
    # Increment the counter
    count += 1
    
    #หาค่า mean
    mean = np.mean(accum_data)
     #update chart title
    plt.title('Random Integers (mean = ' + '{:.2f}'.format(mean) + ')', color='blue')
    
    # Draw the Line chart
    plt.plot(x, y, color='darkgreen', linewidth=1)
    
    display.clear_output(wait=True)
    display.display(plt.gcf())
    
    time.sleep(1)
    
display.clear_output(wait = True)
```

```python
import numpy as np
mpg_ds = np.genfromtxt(
    'mpg.csv', delimiter=',',names=True, dtype=None, encoding='utf-8',
    converters = {
    3: lambda x : float(x.replace('?','0')),
    7: lambda x: x.replace('1','US').replace('2','EU').replace('3','JP')
             }
                        )
print('shape:', mpg_ds.shape, '\n')
print(mpg_ds[0:10])
```

```python
#Create a figure with two subplots in one row
fig = plt.subplots(1,2, figsize=(10,4))

origins, counts = np.unique(mpg_ds['origin'], return_counts=True)
print(origins, '\n', counts)

# Subplot 1
plt.subplot(1, 2, 1)  # สร้าง subplot ที่ตำแหน่ง (1, 1)
plt.bar(origins, counts,color = ['red','green','blue'])
plt.title('Car model by origin')
plt.xlabel('Country')
plt.ylabel('Count')
#add annotation
for x,y in np.stack((origins, counts), axis=1):
    y = int(y)
    plt.annotate(y, (x,y), ha ='center',color = 'white', xytext=(x,y-15))
```

```python
# Subplot 2
plt.subplot(1, 2, 2)  # สร้าง subplot ที่ตำแหน่ง (1, 1)
plt.title('Car model by origin')
plt.pie( counts, labels= origins, autopct ='%1.1f%%',
        colors = ['red','green','blue'],
        shadow=True, explode=(0.2,0,0)
       )
plt.show()
```
